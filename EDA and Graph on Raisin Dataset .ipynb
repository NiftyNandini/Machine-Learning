The different visualizations reveal key characteristics:

Besni raisins tend to be larger overall, with greater Area and ConvexArea. They also have longer axes, making them more spread out.

Kecimen raisins tend to have smaller and more compact shapes but are more eccentric (less circular).

# ALGORITHM

## KNN 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X=df.drop('Class', axis=1) #features
y=df['Class'] #target

#standardize the features
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

#spilt the data
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=42)

#initialize the knn model
knn=KNeighborsClassifier(n_neighbors=5)
#train the model
knn.fit(X_train,y_train)
#make predictions
y_pred=knn.predict(X_test)

#evaluate the accuracy
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy:{accuracy:.2f}')
#classification report
print(classification_report(y_test,y_pred))
#confusion matrix
print(confusion_matrix(y_test,y_pred))
```

## SVM 

```python
from sklearn.svm import SVC

# Initialize the SVM model
svm = SVC(kernel='linear')  

# Train the model
svm.fit(X_train, y_train)

# Make predictions
y_pred = svm.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

SVM algorithm suits best as compare to all other algorithm

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
lr = LogisticRegression(random_state=42)

# Train the model
lr.fit(X_train, y_train)

# Make predictions
y_pred = lr.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)  

# Train the model
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree model
dt = DecisionTreeClassifier(random_state=42)

# Train the model
dt.fit(X_train, y_train)

# Make predictions
y_pred = dt.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# Train a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Create a DataFrame to compare actual vs predicted
results_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

# Identifying correct and incorrect predictions
results_df['Correct'] = results_df['Actual'] == results_df['Predicted']  # True for correct, False for incorrect

# Classes in the dataset (Assuming binary class 0 and 1)
classes = df['Class'].unique()

# Loop over each class and count the correct and wrong predictions
for cls in classes:
    # Filter predictions for the current class
    true_class = results_df[results_df['Actual'] == cls]
    
    # Total correct predictions for this class
    total_true = true_class[true_class['Correct'] == True].shape[0]
    
    # Total wrong predictions for this class
    total_wrong = true_class[true_class['Correct'] == False].shape[0]
    
    print(f"Class '{cls}':")
    print(f"  Total True Predictions: {total_true}")
    print(f"  Total Wrong Predictions: {total_wrong}")
    print()
```

Through EDA and model testing, we conclude that SVM was the best-suited model for classifying raisins with 85% accuracy.
